```python
import numpy as np
import matplotlib.pyplot as plt


#функция, считающая производную
def diff(t,f):
  n=len(t) #общее число точек исходной функции (определение длины вектора t)
  res=np.empty(n) #массив под значения производной (длина n)
  for k in range(n-1):
    res[k] = (f[k+1]-f[k])/(t[k+1]-t[k])
  return res


#функция, считающая интеграл
def integral(t,f):
  n=len(t)
  res=np.empty(n)
  s=0
  for i in range (n-1):
    res[i]=f[i]*(t[i+1]-t[i])
    s=s+res[i]
  print('Интеграл функции = {0:.5f}'.format(s))
  return res


fs=10000  #частота дискретизации
h= 1/fs   #шаг по оси t


#значение интервала и функция
t=np.arange(-0.05,0,h)
f=2*np.sin(t*314)-0.5*np.sin(t*314/4)
#f=t**2   #квадратичная функция
#f=t+1    #линейная функция


#график исходной функции
plt.figure(1)
plt.plot(t,f)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.title('График исходной функции')


#вызов функции для расчета производной
df = diff(t,f)


#график производной функции
plt.figure(2)
plt.plot(t,df)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.title('График производной функции')


#вызов функции для расчета интеграла
dtf = integral(t,f)
```
