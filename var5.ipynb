```python
import pandas as pd  #     
import numpy as np   #   
import matplotlib.pyplot as plt #  
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

#      
def rmse(stationA,stationB,stationC, stationD, forecast):
 stationA, forecast = np.array(stationA), np.array(forecast)
 stationB, forecast = np.array(stationB), np.array(forecast)
 stationC, forecast = np.array(stationC), np.array(forecast)
 stationD, forecast = np.array(stationD), np.array(forecast)
 return [np.sqrt(np.square(np.subtract(stationA,forecast)).mean()),
         np.sqrt(np.square(np.subtract(stationB,forecast)).mean()),
         np.sqrt(np.square(np.subtract(stationC,forecast)).mean()),
         np.sqrt(np.square(np.subtract(stationD,forecast)).mean())]

# 
url = "https://drive.google.com/uc?export=download&id=1R-bDTTCsLjzy9NZG3bkzzeNnFyBuupGK"
input_df = pd.read_csv(url, sep = ';')

#  ,       
input_df.head()
input_df.tail()

#   
stationA = input_df.iloc[:2035 , 1].values
stationB = input_df.iloc[:2035 , 2].values
stationC = input_df.iloc[:2035 , 3].values
stationD = input_df.iloc[:2035 , 4].values
forecast = input_df.iloc[:2035 , 0].values

#     
output_notebook()
a = figure()
b = figure()
c = figure()
d = figure()
#p.line(np.arange(len(forecast)), stationA, line_width = 2, color = 'orange')
#p.line(np.arange(len(forecast)), forecast, line_width = 2, color = 'green')
a.line(np.arange(len(forecast)), np.abs(forecast-stationA), line_width = 1, color = 'red')
b.line(np.arange(len(forecast)), np.abs(forecast-stationB), line_width = 1, color = 'red')
c.line(np.arange(len(forecast)), np.abs(forecast-stationC), line_width = 1, color = 'red')
d.line(np.arange(len(forecast)), np.abs(forecast-stationD), line_width = 1, color = 'red')

#  
a.title.text = '    1'
b.title.text = '    2'
c.title.text = '    3'
d.title.text = '    4'

show(a)
show(b)
show(c)
show(d)

print('Root Mean Squared Error:')
rmse(stationA, stationB, stationC, stationD, forecast)

#       csv    Excel 
data_frame = pd.DataFrame(np.abs(forecast-stationA))
data_frame.to_csv('errorA.csv', index = False, header = ['forecast-station-1'])
data_frame = pd.DataFrame(np.abs(forecast-stationB))
data_frame.to_csv('errorB.csv', index = False, header = ['forecast-station-2'])
data_frame = pd.DataFrame(np.abs(forecast-stationC))
data_frame.to_csv('errorC.csv', index = False, header = ['forecast-station-3'])
data_frame = pd.DataFrame(np.abs(forecast-stationD))
data_frame.to_csv('errorD.csv', index = False, header = ['forecast-station-4'])
```
